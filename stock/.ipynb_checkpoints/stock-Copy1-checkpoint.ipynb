```python
import akshare as ak
import tushare as ts
from datetime import datetime,timedelta
import pandas as pd
from pprint import pprint
```

```python
def get_ma_gt(df,st,en):
    buf = df.loc[st:en].diff().fillna(0)
    res = 0
    for i in df.columns:
        add = buf[buf[i]>0].ma5.count()
        if i in ['ma10','ma20'] and add < 4:
            res = 0;break
        if add >= add_day :
            res+=1 
    return res >= line_count
```

```python
df = ts.get_hist_data('600496') #一次性获取全部日k线数据
df.sort_index(inplace=True)
df['ma30'] = df['close'].rolling(30,min_periods=1).mean()
df['ma60'] = df['close'].rolling(60,min_periods=1).mean()
```

```python
year = '2018-05-01'
df2 = df[col][year:]
```

```python
# col = ['ma5','ma10','ma20','ma30',]#'ma60'
col = ['ma5','ma10','ma20','ma30','ma60']
var = 0.05 #方查系数
add_day = 6 #增长天数
line_count = 2 #均线增长数量
min_date = df2.var(axis=1).sort_values()
min_date = min_date[min_date < var ]
min_date.sort_index()
```

```python
res = []
for date in min_date.index:
    _date = datetime.strptime(date,'%Y-%m-%d')
    st = (_date  + timedelta(days=-12)).date().strftime('%Y-%m-%d')
    en = (_date  + timedelta(days=10)).date().strftime('%Y-%m-%d')
    if get_ma_gt(df2,st,str(_date)):
        res.append(str(date))
```
